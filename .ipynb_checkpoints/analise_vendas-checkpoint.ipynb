## Análise de conversões

```python
df['lead_acesso'] = df['leads_gerados']/df['acessos_site']
df['carrinho_leads'] = df['acesso_carrinho']/df['leads_gerados']
df['vendas_carrinho'] = df['vendas']/df['acesso_carrinho']
df.head()
```

```python
df_copy2 = df.copy()
df_copy2.index = df_copy2['data'] 
df_copy2.head()
```

```python
df_conversion = df_copy2.resample('M').agg({'lead_acesso':'mean', 'carrinho_leads':'mean', 'vendas_carrinho':'mean', 'gasto_midia':'mean'})
df_conversion = df_conversion.reset_index()
df_conversion
```

```python
df_conversion.describe()
```

```python
# Gasto total menor em 03/2020 porém com média mais alta?
# As conversões se mantém relativamente estáveis ao longo do tempo
```

```python
plt.figure()

plt.figure(figsize=(16, 4))
sns.lineplot(x=df_conversion.data, y=df_conversion.lead_acesso, data = df_conversion)
plt.title('Relação entre acessos e leads gerados')

plt.figure(figsize=(16, 4))
sns.lineplot(x=df_conversion.data, y=df_conversion.carrinho_leads, data = df_conversion)
plt.title('Relação entre leads e acesso ao carrinho')

plt.figure(figsize=(16, 4))
sns.lineplot(x=df_conversion.data, y=df_conversion.vendas_carrinho, data = df_conversion)
plt.title('Relação entre acesso ao carrinho e vendas')
```

## Modelo - Arima

```python
# dataset treino e teste
train = df.vendas[0:-8]
test = df.vendas[-8:]
```

```python
train.shape
```

```python
test.shape
```

```python
import statsmodels.api as sm

fig,ax = plt.subplots(2,1,figsize=(15,10))
fig = sm.graphics.tsa.plot_acf(train.diff().dropna(), lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(train.diff().dropna(), lags=50, ax=ax[1])
plt.show()
```

```python
arima = sm.tsa.statespace.SARIMAX(train,order=(12,1,15),seasonal_order=(0,0,0,0),
                                 enforce_stationarity=False, enforce_invertibility=False).fit()
arima.summary()
```

```python
from sklearn.metrics import mean_squared_error
pred = arima.predict(freq=1, start=439, end=446)
```

```python
pd.DataFrame({'test':test,'pred':pred}).plot(figsize=(10, 5))
plt.show()
```

```python
pd.DataFrame({'train': train, 'test': test, 'pred':pred}).plot(figsize=(16, 5))
plt.show()
```

```python
forecast = arima.forecast(steps=14)
```

```python
pd.DataFrame({'train': train, 'test': test, 'forecast': forecast, 'pred':pred}).plot(figsize=(16, 5))
plt.show()
```

```python
# Residuos
res_df = pd.DataFrame(arima.resid,columns=['resid'])
sns.distplot(res_df)
```

```python
from statsmodels.tsa.stattools import acf

def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    
    return({'mape - erro absoluto percentual':mape,
            'me':me,
            'mae - erro absoluto médio': mae, 
            'mpe - erro percentual médio': mpe,
            'rmse -  raiz do erro médio':rmse,
            'corr':corr,
            'minmax':minmax})

forecast_accuracy(pred, test.values)
```

```python
arima.plot_diagnostics(figsize=(15, 12))
plt.show()
```
